Jika berbicara mengenai klasifikasi maka kita akan membicarakan bagaimana mem-plot/memprediksi ke himpunan tertentu. misal memiliki suatu gambar kulit kita dapat mengklasifikasikan apakah termasuk kanker kulit atau bukan atau gambar suatu huruf apakah ia berhuruf a,b,c atau lainnya. Jadi dalam klasifikasi kita akan memplot/memprediksi input kedalam sejumlah kelompok yang terbatas.

Klasifikasi dapat dibagi menjadi linear dan non-linear. maksud dari linear adalah jika independent variables berdimensi n maka pemisah-pemisah kelompok yang berdimensi n-1 akan lurus membagi kelompok-kelompok tersebut. contohnya jika independet variables nya 2 maka pemisahnya adalah garis lurus. jika independent variablesnya 3 maka pemisahnya adalah bidang datar, dan seterusnya. sedangka untuk yang non-linear ia tidak lurus
