## Taking care of missing data

```
from sklearn.impute import SimpleImputer as si
imputer = si(missing_values=np.nan,strategy="mean")
imputer.fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers=[('encoding',OneHotEncoder(),[0])],remainder='passthrough')
x=ct.fit_transform(x)
print(x)
```

Setiap model machine learning hanya bisa memproses data numeric. oleh karena itu jika ada data label (string)/categorical perlu diubah menjadi data numerik. ada 2 cara


1.   Ordinal encoding. dimana feature string akan diubah menjadi integer. setiap nilai variabel direpresentasikan 1 integer. biasanya digunakan jika feature memiliki urutan.
2.   One hot encoding. Digunakan untuk variabel yang tidak memiliki urutan (ranking), seperti nama-nama negara. setiap nilai variabel string, akan direpresentasikan biner. jumlah bit dari biner ini bergantung seberapa banyak nilai yang berbeda yang ditampung variabel tersebut. setiap 1 nilai string akan direpresentasikan 1 bit biner bernilai 1, dan sisa bit biner bernilai 0

pada one hot encoding disini karena kita ingin mengubah country, yang mengandung string dibanding numerik, oleh karena itu harus diubah menjadi numerik. dan kita tahu bahwa negara itu tidak memiliki urutan yang berarti. contoh string yang memiliki urutan adalah 'first', 'second', 'third'.

bisa kita lihat bahwa country memiliki 3 nilai yang berbeda yaitu 'france', 'spain', 'germany'. oleh karena itu kita membutuhkan 3 bit one hot encoding.
karena kita memerlukan 3 bit maka kita memerlukan 3 kolom baru (dummy variables) untuk menyimpan nilai nilai ini, hal ini pula yang mendorong untuk menggunakan columntransformer mengbah 1 kolom menjadi 3 kolom. setelah terjadi transformasi, bisa kita lihat france direpresentasikan 100, spain 001 dan germany 010.




### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
print(y)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x_train[:,3:])
x_train[:,3:] = sc.transform(x_train[:,3:])
print(x_train)
```

Scaling pada Machine Learning ada 2 yaitu
1. Normalization/Min-Max scaling
dimana hasil scaling-nya dalam range [0,1]. rumusnya adalah
$x' = \frac{x-min(x)}{max(x)-min(x)}$

2. Standardization scaling dimana hasil scalingnya memiliki rata-rata 0 dan variansi 1. Biasanya digunakan jika dataset mengikuti fungsi gaussian. Dapat diaplikasikan pada setiap projek. rumusnya
$x' = \frac{x-\mu}{Ïƒ}$
