# Simple Linear Regression

Simple Linear Regression pada dasarnya untuk memprediksi nilai numerik output dari input 1 variabel.
semua linear regression akan berhasil jika hubungan input-output-nya adalah linear

$y=b_0+b_1{.}x$
yang akan dicari adalah nilai b0 dan b1 sehingga nilai error E adalah minimal
${\displaystyle E(b)={\frac {1}{2N}}\sum _{i=1}^{N}(y_{b}(x^{i})-t^{i})^{2}}$ dengan t adalah nilai sesungguhnya/seharusnya dari y 

E tidak lain adalah variance

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

method fit akan menyimpan variabel koefisien dan konstanta linear regression

## Predicting the Test set results

```python
lr.predict(X_test)
```

Akan memprediksi dari garis regresi yang dibuat

## Visualising the Training set results

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,lr.predict(X_train))
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test,y_test,color='red')
plt.plot(X_train,lr.predict(X_train))
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
print("Error",y_test-lr.predict(X_test))
```

```python
#print konstanta dan koefsien
#lr.coef_ adalah list
#destructing list dengan jumlah element 1
koef, = lr.coef_
print("Konstanta : ",lr.intercept_," Koefisien : ",koef)
```
